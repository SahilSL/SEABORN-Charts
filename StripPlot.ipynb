```python
It is basically a scatter plot that differentiate different categories. They are considered a good alternative to a box plot or a violin plot
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
tip=sns.load_dataset("")
sns.stripplot(x="",y="",data=tip)
plt.show()
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.stripplot(x="day", y="total_bill", data=var)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="brg")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="brg", linewidth=1)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="brg", linewidth=1, edgecolor="g")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="brg", linewidth=1, edgecolor="g", jitter=5)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="brg", linewidth=1, edgecolor="g", size=10)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="brg", linewidth=1, edgecolor="g", size=6, marker="*")
plt.show()
```
