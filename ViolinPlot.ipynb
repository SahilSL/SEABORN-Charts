```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.set(style='whitegrid')
f=sns.load_dataset("fmri")
sns.violinplot(x="",y="",data=f)
```

```python
Violin plot plays a similar role as a box and whisker plot. It distribution of quantitative data across several levels of onw or more categorical variables such that those distribution can be compared.
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day", y="total_bill", data=var)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, hue="time")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, linewidth=5)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, linewidth=3, palette="brg")
plt.show()
```
