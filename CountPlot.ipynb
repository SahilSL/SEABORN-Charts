```python
var = sns.load_dataset("tips")
var
```

```python
sns.countplot(x='total_bill', data=var)
plt.show()
```

```python
sns.countplot(x='sex', data=var)
plt.show()
```

```python
sns.countplot(x='day', data=var)
plt.show()
```

```python
sns.countplot(x='sex', data=var, hue="smoker")
plt.show()
```

```python
sns.countplot(y='sex', data=var, hue="smoker")
plt.show()
```

```python
sns.countplot(y='sex', data=var, hue="smoker", palette="ocean_r")
plt.show()
```

```python
sns.countplot(y='sex', data=var, hue="smoker", color="g")
plt.show()
```
