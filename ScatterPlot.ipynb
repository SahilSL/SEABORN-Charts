```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=sns.load_dataset("")
sns.scatterplot(x="",y="",data=data)
plt.show()
```

```python
var=sns.load_dataset("penguins").head(20)
var
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm", data=var)
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm", data=var, hue="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm", data=var, hue="sex", size="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm", data=var, hue="sex", size="sex", palette="gist_ncar")
plt.show()
```

```python
m={"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm", data=var, hue="sex", size="sex", palette="gist_ncar", markers=m)
plt.show()
```
